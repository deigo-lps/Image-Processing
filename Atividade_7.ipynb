Dada a imagem à direita, desenvolva um método capaz de
segmentar as regiões celulares “circulares” (azul/preta).
Descreva cada etapa utilizada para caracterizar o método.
Em seguida, considere algumas regiões de controle e
calcule as taxas de acerto e erro do método. O programa
deve fornecer como saída uma imagem com as regiões
circulares segmentadas e as taxas obtidas.

LF.jpg

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray
from sklearn.metrics import *
image =  io.imread('./LF.jpg')
image = rgb2gray(image)
thresholds = threshold_otsu(image)
image = image > thresholds
plt.subplot(1, 2, 1)

image_guess = io.imread('./guess.jpg')
image_guess = rgb2gray(image_guess)
plt.imshow(image, cmap='gray', interpolation='nearest')
plt.imshow(image_guess,cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(image,cmap='gray')
print("Erro médio", mean_squared_error(image_guess,image, squared=False ))
print("Erro médio absoluto", mean_absolute_error(image_guess,image))
print("Erro médio quadráitco", mean_squared_error(image_guess,image ))
```
